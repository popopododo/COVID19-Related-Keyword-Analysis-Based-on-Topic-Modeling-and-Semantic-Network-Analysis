# Naver Blog Scraper

```python
from selenium import webdriver
from konlpy.tag import Okt
from nltk import Text
from matplotlib import font_manager, rc
from wordcloud import WordCloud
 
import matplotlib.pyplot as plt
import time
import pandas as pd
```

```python
path = "chromedriver.exe" # Put the location of your webdriver.exe
num=0
blogdict = { 'no':[],
            
           'text':[]
           }
test = pd.DataFrame.from_dict(blogdict)
test.to_csv("Blog_화이자.csv",encoding="utf-8-sig")
driver = webdriver.Chrome(path) 
url_list = [] # Variable for blog url
content_list = "" 
text = "백신" # Keyword (Vaccine, Covid)
 
for i in range(1,201):  # Fetch blog articles from page 1 to 200
    url = 'https://section.blog.naver.com/Search/Post.nhn?pageNo='+ str(i) + '&rangeType=ALL&orderBy=sim&keyword=' + text # url 값 설정
    driver.get(url)
    time.sleep(0.5) # Preventing sleep for errors
 
    for j in range(1, 5):
        titles = driver.find_element_by_xpath('/html/body/ui-view/div/main/div/div/section/div[2]/div['+str(j)+']/div/div[1]/div[1]/a[1]')
        title = titles.get_attribute('href')
        url_list.append(title)
print(url_list)
print("url 수집 끝, 해당 url 데이터 크롤링")
 
for url in url_list: 
    driver.get(url)
 
    driver.switch_to.frame('mainFrame')
    dateoverlay = '#SE-30d1a1f0-2906-11ec-8d2f-1727c85e413d > div.se-component-content > div > div.blog2_container > span.se_publishDate.pcol2'
    overlays = ".se-component.se-text.se-l-default" # Scrape Blog Articles
    date = driver.find_elements_by_css_selector(dateoverlay)
    contents = driver.find_elements_by_css_selector(overlays)
    
    
    content_list=""
    for content in contents:
        content_list = content_list + content.text 
    num+=1
    blogdict['no'].append(num) 
    blogdict['text'].append(content_list)
    blogdict['date'].append(date)
       
      
test = pd.DataFrame.from_dict(blogdict)
test.to_csv("Blog_화이자1-200.csv",encoding="utf-8-sig")
print(blogdict)
```
